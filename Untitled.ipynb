```python
import pandas as pd
from sklearn.linear_model import LinearRegression


df=pd.read_csv('../used_car_dataset.csv')
```

```python
df.head()
```

```python
df['B_c'] = df['Brand'].replace({
    'Honda':1, 'Toyota':2, 'Volkswagen':3, 'Maruti Suzuki':4, 'BMW':5, 'Ford':6,
    'Kia':7, 'Mercedes-Benz':8, 'Hyundai':9, 'Audi':10, 'Renault':11, 'MG':11,
    'Volvo':12, 'Skoda':13, 'Tata':14, 'Mahindra':15, 'Mini':16, 'Land Rover':17,
    'Jeep':18, 'Chevrolet':18, 'Jaguar':19, 'Fiat':20, 'Aston Martin':21, 'Porsche':22,
    'Nissan':23, 'Force':24, 'Mitsubishi':25, 'Lexus':26, 'Isuzu':27, 'Datsun':28,
    'Ambassador':29, 'Rolls-Royce':30, 'ICML':31, 'Bajaj':32, 'Opel':33, 'Ashok':34,
    'Bentley':35, 'Ssangyong':36, 'Maserati':37
})

```

```python
df['Transmission'] = df['Transmission'].replace({'Manual':1, 'Automatic':2})
df['FuelType'] = df['FuelType'].replace({'Petrol':1, 'Diesel':2, 'Hybrid/CNG':3})
df['Owner'] = df['Owner'].replace({'first':1, 'second':2})

```

```python
# Clean and convert kmDriven to numeric
df['km_d'] = df['kmDriven'].str.replace(',', '').str.extract('(\d+)').astype(float)
```

```python
# Clean and convert AskPrice to numeric
df['AskPrice'] = df['AskPrice'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
# Prepare features and target
x = df[['B_c', 'Year', 'km_d', 'Transmission', 'Owner', 'FuelType']]
y = df['AskPrice']

# Fill missing values with mean
x_filled = x.fillna(x.mean())

# Train the model
model = LinearRegression()
model.fit(x_filled, y)
```

```python
def predict_car_price():
    # Brand mapping dictionary
    brand_mapping = {
        'Honda':1, 'Toyota':2, 'Volkswagen':3, 'Maruti Suzuki':4, 'BMW':5, 'Ford':6,
        'Kia':7, 'Mercedes-Benz':8, 'Hyundai':9, 'Audi':10, 'Renault':11, 'MG':11,
        'Volvo':12, 'Skoda':13, 'Tata':14, 'Mahindra':15, 'Mini':16, 'Land Rover':17,
        'Jeep':18, 'Chevrolet':18, 'Jaguar':19, 'Fiat':20, 'Aston Martin':21, 'Porsche':22,
        'Nissan':23, 'Force':24, 'Mitsubishi':25, 'Lexus':26, 'Isuzu':27, 'Datsun':28,
        'Ambassador':29, 'Rolls-Royce':30, 'ICML':31, 'Bajaj':32, 'Opel':33, 'Ashok':34,
        'Bentley':35, 'Ssangyong':36, 'Maserati':37
    }
    
    # Fuel type mapping
    fuel_map = {
        'petrol': 1,
        'diesel': 2,
        'hybrid': 3,
        'cng': 3,
        'hybrid/cng': 3
    }
    
    # Get user input
    print("Enter car details:")
    brand_name = input("Brand: ").title()  # Convert to title case for matching
    year = int(input("Manufacturing Year: "))
    km_d = int(input("Kilometers Driven: "))
    transmission = input("Transmission Type (Manual/Automatic): ").lower()
    owner = int(input("Owner (1 for first, 2 for second): "))
    fuel_type = input("Fuel Type (Petrol/Diesel/Hybrid/CNG): ").lower()
    
    # Convert categorical inputs to numerical codes
    brand_code = brand_mapping.get(brand_name, 0)  # Default to 0 if brand not found
    transmission_code = 1 if transmission == 'manual' else 2
    fuel_code = fuel_map.get(fuel_type, 1)  # Default to petrol if not found
    
    # Create input array for prediction
    input_data = [[brand_code, year, km_d, transmission_code, owner, fuel_code]]
    
    # Make prediction
    predicted_price = model.predict(input_data)[0]
    
    print(f"\nPredicted Price: ₹{predicted_price:,.2f}")

# Run the prediction function
predict_car_price()
```
